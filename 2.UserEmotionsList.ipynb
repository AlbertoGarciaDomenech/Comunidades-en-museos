# Lista de emociones de cada usuario
Añadir al dataset de usuarios que cuadros les ha generado a cada usuario emociones positivas, negativas y mixtas

---

```python
import pandas as pd
```

```python
prado_emotions_data = 'data/Prado_emotions.csv'
prado_emotions = pd.read_csv(prado_emotions_data)
prado_emotions.head(5)
```

```python
### Crear dataset con usuarios y listas de sus cuadros positivos, negativos y mixed 
users_emotions_list = {}
i = 1

for id in list(prado_emotions.userId.unique()):
    user = prado_emotions[prado_emotions['userId'] == id]
    
    pos_artworks = list(user[user['Polarity'] == 'positive'].artworkId)
    neg_artworks = list(user[user['Polarity'] == 'negative'].artworkId)
    mix_artworks = list(user[user['Polarity'] == 'mixed'].artworkId)
    
    users_emotions_list.update({i: [id, pos_artworks, neg_artworks, mix_artworks]})
    i += 1
    
users_emotions_df = pd.DataFrame.from_dict(users_emotions_list, orient='index', columns=['userId', 'positive', 'negative', 'mixed'])
users_emotions_df = users_emotions_df.sort_values(by=['userId'])
users_emotions_df = users_emotions_df.set_index(pd.Index([i for i in range(0,len(users_emotions_df))]))
```

```python
users_emotions_df
```

```python
prado_users = pd.read_csv('data/Prado_users.csv')

pos = users_emotions_df['positive']
neg = users_emotions_df['negative']
mix = users_emotions_df['mixed']

prado_users = prado_users.join(pos)
prado_users = prado_users.join(neg)
prado_users = prado_users.join(mix)

prado_users.to_csv('data/Prado_users_emotions.csv', index=0)
```

```python
prado_users_scaled = pd.read_csv('data/Prado_users_scaled.csv')

pos = users_emotions_df['positive']
neg = users_emotions_df['negative']
mix = users_emotions_df['mixed']

prado_users_scaled = prado_users_scaled.join(pos)
prado_users_scaled = prado_users_scaled.join(neg)
prado_users_scaled = prado_users_scaled.join(mix)

prado_users_scaled.to_csv('data/Prado_users_emotions_scaled.csv', index=0)
```
