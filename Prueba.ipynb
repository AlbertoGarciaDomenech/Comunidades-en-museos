```python
# REDUCIR A PROCEDENTE DE ESPAÑA O NO
# ya que el dataset de prado_users tiene pais vamos a utilizar el dataset countries of the world para identificar las regiones de las que proceden
countries_data = "data/countries_of_the_world.csv"
countries = pd.read_csv(countries_data)
countries = countries.filter(items = ['Country', 'Region'])
countries.head()
```

```python
# eliminar el espacio en blanco despues de cada nombre de pais
for i in range(0, countries.shape[0] - 1):
    countries.loc[i] = countries.loc[i].replace(countries.loc[i]['Country'], countries.loc[i]['Country'][:-1])
countries
```

```python
# sustituimos cada pais de procedencia de prado_users por un numero q corresponda con su region
# ver cuantas regiones hay
num_regions = 0
for i in countries['Region'].unique():
    num_regions += 1
    countries = countries.replace(i,num_regions)
```

```python
# formatear los nombres de USA y Argentina
prado_users = prado_users.replace('USA','United States')
prado_users = prado_users.replace('Argenitina','Argentina')
# reemplazar los paises por el numero de regiones
for country in prado_users['country'].unique():
    if country == 'Other':
        prado_users = prado_users.replace(country,0) #sustituir por 0 los usuarios que no sabemos de donde son
    else:
        prado_users = prado_users.replace(country,countries.loc[countries['Country'] == country]['Region'].to_list()[0])
```

```python
# age
# formatear los <12 y >70
prado_users = prado_users.replace('<12', '0-12')
prado_users = prado_users.replace('>70', '70-100')

age_index = 1

for age_range in prado_users.sort_values(by=['age'])['age'].unique():
    prado_users = prado_users.replace(age_range, age_index)
    age_index += 1
age_index -=1
```

```python
# gender
prado_users = prado_users.replace('female', 0)
prado_users = prado_users.replace('male', 1)
```

```python
emotions = emotions.replace('positive', 1)
emotions = emotions.replace('negative', -1)
emotions = emotions.replace('mixed', 0)
emotions[emotions['userId'] == 8]
```

```python
aux = prado_users.copy()
aux
```

```python
for artwork in emotions['artworkId'].unique():
#     print('ARTWORK: ', artwork)
    polarities = []
    for user in prado_users['userId']:
#         print('USER: ', user)
        
        emo = emotions[emotions['userId'] == user][emotions['artworkId'] == artwork]
#         print(emo['Polarity'])
        
        if emo.empty:
            pol = 0 #No opinion
        elif np.sum(emo['Polarity']) > 0:
            pol = 1 #Positive
        elif np.sum(emo['Polarity']) < 0:
            pol = 2 #Negative
        else:
            pol = 3 #Mixed

#         print(pol)
        polarities.append(pol)
    
#     print('\n', polarities)
    aux[artwork] = polarities
```

```python
#datos personales: edad/genero/pais
def similarityPersonal(u1, u2):
#     u1 = prado_users.loc[prado_users['userId'] == user1] #usuario1
#     u2 = prado_users.loc[prado_users['userId'] == user2] #usuario2
    pesoPais = 0.3
#     VER SI PODEMOS ORDENAR LAS REGIONES COMO EDADES
    simPais = (u1['country'].to_list()[0] == u2['country'].to_list()[0])
    pesoEdad = 0.5
#     HAY QUE ORDENAR LAS EDADES: en vez de llamar 1 al rango 55-64 llamarlo al 0-12
#  fSimiliaritud n1 y n2 = 1 - ((1/age_index)*abs(n1-n2))
#    [HECHO]
    simEdad = 1 - ((1/(age_index-1)) * abs(u1['age'].to_list()[0]-u2['age'].to_list()[0]))
#     simEdad = (u1['age'].to_list()[0] == u2['age'].to_list()[0])
    pesoGen = 0.2
    simGen = (u1['gender'].to_list()[0] == u2['gender'].to_list()[0])

    print(simPais*pesoPais + simEdad*pesoEdad + simGen*pesoGen)   
```
