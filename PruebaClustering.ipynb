```python
import pandas as pd
import numpy as np
from sklearn_extra.cluster import KMedoids
```

```python
u = pd.read_csv('data/usersSim.csv')
u = u.drop(columns = ['Unnamed: 0'])
u
```

```python
kmedoids = KMedoids(init='random', n_clusters= 10, metric="precomputed", random_state=123,max_iter=100000)
kmedoids.fit(u)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aux = pd.read_csv("data/Prado_users.csv")
aux.drop('userId', axis=1, inplace=True) 
aux['cluster'] = kmedoids.labels_
```

```python
aux
```

```python
sns.pairplot(aux,hue = 'cluster')
plt.show()
```
