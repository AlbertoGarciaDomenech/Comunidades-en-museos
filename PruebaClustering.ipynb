```python
import pandas as pd
import numpy as np
from sklearn_extra.cluster import KMedoids
import networkx as nx
```

```python
u = pd.read_csv('data/usersSim.csv', index_col=0)
u
```

```python
kmedoids = KMedoids(init='random', n_clusters= 10, metric="precomputed",max_iter=100000)
kmedoids.fit(u)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kmedoids.labels_
```

```python
np.unique(kmedoids.labels_)
```

```python
aux = pd.read_csv("data/Prado_users.csv")
aux.drop('userId', axis=1, inplace=True) 
aux['cluster'] = kmedoids.labels_
```

```python
aux.to_csv('data/usersSimCLUSTER.csv')
```

```python
g = nx.Graph()

for i in range(len(aux)): #meter las personas en el grafo
    g.add_node(i)
    nx.set_node_attributes(g, aux.loc[i][0], "edad")
    nx.set_node_attributes(g, aux.loc[i][1], "genero")
    nx.set_node_attributes(g, aux.loc[i][2], "pais")
    nx.set_node_attributes(g, aux.loc[i][3], "cluster")
    print(aux.loc[i][3])
    
    
for i in range(len(u)): #add las aristas entre personas -> similitudes
    for j in range(len(u)):
        if j != i:
            g.add_edge(i,j,weight = u.loc[i][j])
```
